```python
# 3
def add(a, b):
    return a + b

print(add(5, 3))
```

```python
# 4
def subtract(a, b):
    return a-b

print(subtract(8, 5))
```

```python
# 5
def multiply(a, b):
    return a * b

print(multiply(6,6))
```

```python
# 6
def divide(a, b):
    return a/b

print(divide(100,2))
```

```python
# 7
def is_even(number):
    return number % 2 == 0

print(is_even(4))   # True
print(is_even(5))   # False
```

```python
# 8
def find_max(a, b):
    return a if a > b else b

print(find_max(1150, 2500))

```

```python
# 10
def cube(number):
    return number **3

print(cube(6))
```

```python
# 11
def count_letters(word):
    return len(word)

print(count_letters("Sunshine"))
```

```python
# 12
def convert_to_upper(text):
    return text.upper()

print(convert_to_upper("hello world"))
```

```python
# 13
def calculate_area(length, width):
    return length * width

print(calculate_area(15,30))
```

```python
# 14
def average(numbers):
    return sum(numbers) / len (numbers)

print(average([5,10,15,20,25]))
```

```python
# 15
def factorial(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

print(factorial(50))
```
